* 下载波士顿数据集，读取全部506条数据，放在NumPy数组x、y中（x：属性，y：标记）。(5分)

* 使用全部506条数据，实现波士顿房价数据集可视化，如图1所示。(10分)

* 要求用户选择属性，如图2所示，根据用户的选择，输出对应属性的散点图，如图3所示。(5分)

```
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

boston_housing = tf.keras.datasets.boston_housing
(x,y),(_,_) = boston_housing.load_data(test_split=0)

titles = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B-1000','LSTAT','MEDV']

plt.figure(1,(12,12))
for i in range(13):
  plt.subplot(4,4,i+1)
  plt.scatter(x[:,i],y,label=titles[i])
  plt.xlabel(titles[i])
  plt.ylabel("Price:$1000")
  plt.title(str(i)+' '+titles[i]+" - PRICE")
  plt.legend()

plt.tight_layout()
plt.suptitle("The factor with House_Price",x = 0.5 ,y=1.05,fontsize = 20)
plt.show()

for i in range(13):
  print(i+1," -- "+titles[i])
num = int(input("Please select the factor number:"))
plt.figure()
plt.scatter(x[:,num-1],y)
plt.xlabel(titles[num-1])
plt.ylabel("Price:$1000")
plt.title(str(num)+' '+titles[num-1]+" - PRICE")
plt.tight_layout()
plt.show()
```
